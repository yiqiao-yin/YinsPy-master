### Libraries

```python
import numpy as np
import pandas as pd
```

### Create Data

```python
n = 500
N = n + 1000
cutoff = round(n/N, 1)
p = 30
X = np.reshape(np.random.randint(0, 2, n*p), (n, p))
print(X)
```

```python
pd.DataFrame(X).head(3)
```

```python
X = pd.DataFrame(X)
X.describe()
```

### Create Underlying Model

```python
y1 = np.mod(X.iloc[:, 1] + X.iloc[:, 2], 2)
y2 = np.mod(X.iloc[:, 3] + X.iloc[:, 4] + X.iloc[:, 5], 2)
```

```python
y = np.where(np.random.randint(0, 2, n) == 1, y1, y2)
print(y.mean())
print(y)
```

```python
sample_data = pd.concat([X.iloc[:, [1,2]], pd.DataFrame(y)], axis=1)
sample_data.columns = ['1', '2', 'y']
sample_data.head()
```

### Theoretical Accuracy

```python
np.mod(X.iloc[:, [3,4,5]].sum(axis=1), 2)
```

```python
LossL1 = y - np.mod(sample_data.iloc[:, 1] + sample_data.iloc[:, 2], 2)
CHK = np.where(LossL1 != 0, 0, 1)
print('1st Round:', CHK)
print('Accuracy:', CHK.mean())

LossL2 = np.where(LossL1 != 0, LossL1 - np.mod(X.iloc[:, [3,4,5]].sum(axis=1), 2), LossL1)
CHK = np.where(LossL2 == 0, 1, 0)
print('2nd Round:', CHK)
print('Final Accuracy:', CHK.mean())
```
